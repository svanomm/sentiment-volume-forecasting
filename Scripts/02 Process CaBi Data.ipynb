```python
import pandas as pd
import os
```

```python
# Get the list of CSV files
download_folder = "../Data/Raw"
csv_files = [f'{download_folder}/{f}' for f in os.listdir(download_folder) if f.endswith('.csv')]
```

```python
csv_files
```

```python
# Stack the CSV files into one dataframe
df = pd.concat([pd.read_csv(f) for f in csv_files], ignore_index=True)
```

```python
df
```

```python
# How many origin-destination pairs?
# Create column OD which concatenates start_lat, start_lng, end_lat, end_lng
df['OD'] = df['start_lat'].astype(str) + df['start_lng'].astype(str) + df['end_lat'].astype(str) + df['end_lng'].astype(str)
# Count the number of unique origin-destination pairs
num_od_pairs = df['OD'].nunique()
```

```python
num_od_pairs
```

```python
df[df['member_casual']=='casual']
```

```python
df['ride_time'] = pd.to_datetime(df['ended_at']) - pd.to_datetime(df['started_at'])
# Convert ride_time to minutes
df['ride_time'] = df['ride_time'].dt.total_seconds() / 60

df['ride_time'].describe()
```

```python
# CLEANING STEPS

# Drop 0 or negative ride times
df = df[df['ride_time'] > 0]

# Drop suspiciously long ride times
df = df[df['ride_time'] < 60]

# Drop rides that start and end at the same location
df = df[df['start_station_id'] != df['end_station_id']]

# Remove member rides (because we can't estimate price)
df = df[df['member_casual'] == 'casual']

# Drop missing lat/lon
df = df.dropna(subset=['start_lat', 'start_lng', 'end_lat', 'end_lng'])

# Drop any lat/lon equal to 0
df = df[(df['start_lat'] != 0) & (df['start_lng'] != 0) & (df['end_lat'] != 0) & (df['end_lng'] != 0)]
```
